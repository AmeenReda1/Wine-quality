```python
## This is wine Data analysis project form UCI
import pandas as pd
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
df_red=pd.read_csv('winequality-red.csv')
df_red.head()
```

```python
# add new column in red_wine called color and the value in red
data_color_red=["red"]*len(df_red)
#df_red.insert(13,"Color",data_color_red,True)
#df_red.head()
df_red["color"]=data_color_red
df_red.head()
```

```python
# add new column in white_wine called color and the value in white
data_color_white=["white"]*len(df_white)

df_white["color"]=data_color_white
df_white.head()
```

```python
#merge the white_wine and red_wine
all_wine_data=df_white.append(df_red,ignore_index=True)
all_wine_data
```

```python
# groub by color to know whice type of wine has more quality
all_wine_data.groupby("color",as_index=False).mean()
```

```python
bin_edges = [2.72, 3.11, 3.21, 3.32, 4.01]

all_wine_data["pH category"]=pd.cut(all_wine_data['pH'],bin_edges,labels=['High','Moderately High','Medium','Low'])
all_wine_data
```

```python
all_wine_data.groupby("pH category",as_index=False).mean()
```

```python
high_alcohol=all_wine_data.query('alcohol >= alcohol.median()')
lower_alcohol=all_wine_data.query('alcohol < alcohol.median()')
high_alcohol['quality'].mean()
```

```python
lower_alcohol['quality'].mean()
```

```python
high_sweet_wine=all_wine_data[all_wine_data['residual sugar']>=all_wine_data['residual sugar'].median()]
low_sweet_wine=all_wine_data[all_wine_data['residual sugar']<all_wine_data['residual sugar'].median()]
high_sweet_wine['quality'].mean()
```
